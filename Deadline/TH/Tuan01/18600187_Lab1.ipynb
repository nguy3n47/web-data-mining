```python
# import requests module
import requests
import re
from urllib.request import urlopen
from google.colab import drive


drive.mount('/content/drive')
urls = urlopen("https://www.hcmus.edu.vn/")
text = urls.read()
data = text.decode('utf8')
links = re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', data)
myList = list(set(links))

for temp in myList:
  t=requests.get(temp)
  if t.status_code!=200:
    myList.remove(temp)

fop=open('/content/drive/My Drive/18600187.txt','w')
for link in myList:
  fop.write(link)
  fop.write('\n')
fop.close()
print(myList)
```
